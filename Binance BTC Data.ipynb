### Identifying if csvs are accessed properly and stored in df_1, df_2 up to df_24 

```python
import pandas as pd

start_date = "2021-04-01"
end_date = "2023-04-01"
date_range = pd.date_range(start=start_date, end=end_date, freq="MS")
data_frames = {}

for i, date in enumerate(date_range):
    filename = f"C:/Users/Harvey/Downloads/BTCUSDT-1m-{date.strftime('%Y-%m')}.csv"
    data_frames[f"df_{i+1}"] = pd.read_csv(filename)
    print(filename)

# assign the data frames to variable names using locals()
for i, key in enumerate(data_frames):
    locals()[f"df_{i+1}"] = data_frames[key]
    print(i)
```
